```
# Q1: Multiplication Table
print("Q1: Multiplication Table")
n = int(input("Enter a number for multiplication table: "))
for i in range(1, 11):
    print(f"{n} x {i} = {n * i}")

print("\n" + "-"*50 + "\n")


# Q2: Fibonacci Series
print("Q2: Fibonacci Series")
terms = int(input("Enter number of terms: "))
a, b = 0, 1
count = 0
while count < terms:
    print(a, end=" ")
    a, b = b, a + b
    count += 1
print()

print("\n" + "-"*50 + "\n")


# Q3: Prime Number Checker
print("Q3: Prime Number Checker")
num = int(input("Enter a number: "))

if num <= 1:
    print(f"{num} is NOT a prime number.")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{num} is a PRIME number.")
    else:
        print(f"{num} is NOT a prime number.")

print("\n" + "-"*50 + "\n")


# Q4: Power Calculation
print("Q4: Power Calculation")
b = float(input("Enter base: "))
e = int(input("Enter exponent: "))
result = b ** e
print(f"{b} raised to the power {e} is {result}")

print("\n" + "-"*50 + "\n")


# Q5: Reverse Number
print("Q5: Reverse Number")
n = int(input("Enter an integer: "))
reverse = 0
temp = n
while temp > 0:
    digit = temp % 10
    reverse = reverse * 10 + digit
    temp //= 10
print(f"Original number: {n}")
print(f"Reversed number: {reverse}")

print("\n" + "-"*50 + "\n")


# Q6: Guess Until Correct
print("Q6: Guess Until Correct")
secret = 7  # You can change the secret number
while True:
    guess = int(input("Guess the secret number: "))
    if guess == secret:
        print(" Correct! You guessed it.")
        break
    else:
        print(" Wrong! Try again.")
```
